```python
first_inning_score=df.loc[df["inning"]==1].groupby(['match_id'])['total_runs'].sum().reset_index()
first_inning_score
```

```python
secoond_inning_score=df.loc[df["inning"]==2].groupby(['match_id'])['total_runs'].sum().reset_index()
secoond_inning_score
```

```python
merged_df= pd.merge(first_inning_score, secoond_inning_score, on='match_id', suffixes=('_first_inning', '_second_inning'))
merged_df
```

```python
merged_df["winner"]=np.where(merged_df["total_runs_first_inning"]>merged_df["total_runs_second_inning"],"first_inning",np.where(merged_df["total_runs_first_inning"]<merged_df["total_runs_second_inning"],"second_inning","TIE"))
```

```python
merged_df
```

```python
merged_df.drop(["total_runs_first_inning","total_runs_second_inning"],axis=1,inplace=True)
```

```python
merged_df
```

```python
final_df=pd.merge(df,merged_df,how="left",on="match_id")
```
